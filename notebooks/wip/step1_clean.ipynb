```python
import pandas as pd
import scipy.stats
```

```python
###### hardcoded
file = '../../data/heatmap/BRCA_fts_by_VALUE.tsv'
datatype = 'N:METH'
out = '../../data/heatmap/BRCA_vals_meth.tsv'
######
```

```python
# Format row:col:val - ft:sample:tarballval
df = pd.read_csv(file, sep='\t', index_col=0)
df = df.drop(['Labels'], axis=1) # rm col

# filter for datatype cols
keep =[]
for c in df.columns:
    if c.startswith(datatype):
        keep.append(c)
print(len(keep), '{} features kept'.format(datatype))
df = df[keep]
df
```
